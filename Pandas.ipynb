SERIES INDEXING

Using .loc (Label-based)

Using .iloc (Position-based)

```python
import pandas as pd
scores = pd.Series([85, 92, 78, 95, 88],index=['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],name="marks")
scores

```

```python
# using .loc (label based)

# single element
print(scores.loc["Alice"])
print()

# multi ele
print(scores.loc[["Bob","Eve"]])
print()

# slicing by labels[inclusive]
print(scores.loc["Charlie":"Eve"])
print()

# Boolean condition
print(scores.loc[scores>90])
```

```python
# Using .iloc (Position-based)

# Single element by position
print(scores.iloc[1])

print()

# multiple
print(scores.iloc[[0,4]])
print()

# Slicing by position (EXCLUSIVE end!)
print(scores.iloc[1:4])
print()
# negative indexing
print(scores.iloc[-4:-2])
print()
# reverse
print(scores.iloc[-4:][::-1])

```

```python
import pandas as pd
# Student grades dataset
data = {
    'Math': [85, 92, 78, 95, 88],
    'Science': [90, 88, 85, 98, 91],
    'English': [88, 85, 92, 90, 87]
}
df = pd.DataFrame(data,
                  index=['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'])
print(df)
df.shape
```

```python
# .loc - Label-based Selection

# Single cell (row_label, column_label)
print(df.loc["Charlie","Science"])

# entire row
print(df.loc["Charlie"])
print()

# entire col
print(df.loc[:,"Science"])
print()
# multiple rows
print(df.loc[["Alice","Eve"]])
print()

# # multiple col
print(df.loc[:,["Math","English"]])

# Subregion (rows + columns)
print(df.loc["Bob":"Diana","Science":"English"])
```
